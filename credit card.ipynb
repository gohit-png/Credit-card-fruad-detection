```python
train_set, test_set = train_test_split(df, test_size = 0.2, random_state = 42)
print(f"rows in train set = {len(train_set)} \nrows in test set = {len(test_set)}\n")
```

```python
 from sklearn.model_selection import StratifiedShuffleSplit 
 sss = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
 for train_index, test_index in sss.split(df, df["Class"]):
     strat_train_set = df.loc[train_index] 
 strat_test_set = df.loc[test_index]
```

```python
X_train = strat_train_set.drop("Class", axis=1)
```

```python
Y_train = strat_train_set["Class"].copy()
```

```python
X_test = strat_test_set.drop("Class", axis=1)
X_test.head()
```

```python
Y_test = strat_test_set["Class"].copy()
Y_test
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

my_pipeline = Pipeline([('std_scaler', StandardScaler())])

pip_tr = my_pipeline.fit_transform(X_train)
pip_tr


```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
#model = SVC()
model = RandomForestClassifier(n_estimators=20)#it gave best predictions
#model = GaussianNB()
#model = LogisticRegression()
model.fit(pip_tr, Y_train)
```

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(model, pip_tr, Y_train, cv=5, scoring = "accuracy")
```

```python
def print_scores(score):
    print("score: ", score)
    print("mean: ", score.mean())
    print("SD :", score.std())
    
    
 
```

```python
print_scores(score)
```

```python
from sklearn.model_selection import cross_val_predict
Y_train_predict = cross_val_predict(model, pip_tr, Y_train, cv=5)

Y_train_predict

```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(Y_train, Y_train_predict)

```

```python
from sklearn.metrics import f1_score, precision_score

print(precision_score(Y_train, Y_train_predict))
print(f1_score(Y_train, Y_train_predict))


```

```python
X_test = strat_test_set.drop('Class', axis=1)
Y_test = strat_test_set['Class'].copy()
pip_test = my_pipeline.fit_transform(X_test)
final_prediction = model.predict(pip_test)
final_prediction
```

```python
print(precision_score(Y_test, final_prediction))
print(f1_score(Y_test, final_prediction))
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,final_prediction))
```
